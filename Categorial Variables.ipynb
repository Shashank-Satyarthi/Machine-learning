```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
plt.xlabel('Mileage')
plt.ylabel('Sell Price')
plt.scatter(df.Mileage,df['Sell Price($)'],c='red',marker='+')
```

```python
plt.xlabel('Age')
plt.ylabel('Sell Price')
plt.scatter(df['Age(yrs)'],df['Sell Price($)'],c='blue',marker='o')
```

# Using pandas to create dummy variable

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
final = merged.drop(['Car Model','Mercedez Benz C class'],axis=1)
final
```

```python
X = final.drop(['Sell Price($)'],axis=1)
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

#### Predict price of a mercedez benz that is 4 yr old with mileage 45000

```python
lr.predict([[4500,4,0,0]])
```

#### Predict price of a BMW X5 that is 7 yr old with mileage 86000

```python
lr.predict([[8600,7,0,1]])
```

#### Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

```python
lr.score(X,y)
```
